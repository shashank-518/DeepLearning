```python
import pandas as pd
from keras.preprocessing.text import one_hot
```

```python
sentences = [
    'The Glass of Milk',
    'The Glass of Water',
    'the Cup of Tea',
    'I am Software Engineer',
    'Your Videos Are Good'
]
```

```python
voc_words =10000
onehot = [one_hot(word,voc_words) for word in sentences]
onehot
```

```python
from keras.models import Sequential
from keras.utils import pad_sequences
from keras.layers import Embedding
```

```python
sent_length = 8
word_len = pad_sequences(onehot , maxlen=sent_length,padding='pre')
word_len
```

```python
dim = 10

model = Sequential()
model.add(Embedding(voc_words,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(word_len)
```
