```python
import pandas as pd
import numpy as np
from keras.datasets import imdb
from keras.models import Sequential,load_model
from keras.layers import Dense,Embedding,SimpleRNN
from keras.preprocessing import sequence
```

```python
word_index = imdb.get_word_index()
reverse_words = {value:key for key,value in word_index.items()}
```

```python
model = load_model('simple_rnn.h5')
```

```python
model.summary()
```

```python
model.get_weights()
```

```python
def decode_review(original):
    return ' '.join([reverse_words.get(i-3,'?') for i in original])
```

```python
def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word,2)+3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review],maxlen=500)
    return padded_review
```

```python
def movieReview(rev):
    processed_input = preprocess_text(rev)

    prediction = model.predict(processed_input)

    sentiment = 'Positive' if prediction[0][0] > 0.5 else 'Negative'

    return sentiment,prediction[0][0]
```

```python
review = "The Movie is fanastic and quite Thrilling"
```

```python
setminent,pred=movieReview(review)
```
